### Data Cleaning

```python
# finding no_guests with negative values
df_bookings[df_bookings.no_guests <= 0]
```

```python
df_bookings.shape
```

```python
# get rid of those negative values
df_bookings = df_bookings[df_bookings.no_guests > 0]
df_bookings.shape
```

```python
# min & max value of revenue generated
df_bookings.revenue_generated.min(), df_bookings.revenue_generated.max()
```

```python
avg, std = df_bookings.revenue_generated.mean(), df_bookings.revenue_generated.std()
avg, std
```

```python
# outlier detection
higher_limit = avg + 3 * std
higher_limit
```

```python
lower_limit = avg - 3 * std
lower_limit
```

```python
# no negative values in revenue_generated col in the dataframe
df_bookings[df_bookings.revenue_generated < 0]
```

```python
# printing outliers (value greater than 3 times std from mean)
df_bookings[df_bookings.revenue_generated > higher_limit]
```

```python
# dataframe with outliers removed
df_bookings = df_bookings[df_bookings.revenue_generated < higher_limit]
df_bookings.shape
```

```python
# check stats for revenue_realized column
df_bookings.revenue_realized.describe()
```

```python
# check for outliers in revenue_realized
avg, std = df_bookings.revenue_realized.mean(), df_bookings.revenue_realized.std()
avg, std
```

```python
higher_limit = avg + 3*std
higher_limit
```

```python
# print higher-limit outliers
df_bookings[df_bookings.revenue_realized > higher_limit] 
```

### Data Transformation - creating new column / normalization / aggregating data

```python
df_agg_bookings.head()
```

```python
# creating a column - occupancy %
df_agg_bookings['occ_pct'] = df_agg_bookings['successful_bookings'] / df_agg_bookings['capacity']
```

```python
df_agg_bookings.head()
```

```python
df_agg_bookings['occ_pct'] = df_agg_bookings['occ_pct'].apply(lambda x: round(x*100,2))
df_agg_bookings.head()
```

### Insights Generation

Q. What is the average occupancy rate in each of the room categories?

```python
df_agg_bookings.groupby('room_category')['occ_pct'].mean().round(2)
```

```python
df_rooms
```

```python
# merge df_agg_bookings with df_rooms 
df = pd.merge(df_agg_bookings, df_rooms, left_on="room_category", right_on="room_id")
df.head()
```

```python
df.groupby('room_class')['occ_pct'].mean().round(2)
```

```python
# dropping 1 column because of repetition
df.drop('room_id', axis=1, inplace=True)
df.head()
```
